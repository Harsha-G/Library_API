```python
api_url = "http://127.0.0.1:8080/"

import requests
```

```python
# API Command to List All Books

command = "list_books/"
api_url+=command

r = requests.get(api_url)
print(r.text)
```

```python
# API Command to Add a new Book entry

data = { "title" : "Cosmos", "author" : "Carl Sagan", "genre" : "SCI" }
command = "add_book/"
api_url+=command

r = requests.post(api_url, data=data)
print(r.text)
```
